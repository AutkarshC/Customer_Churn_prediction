```python
import pandas as pd
import numpy as np 
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Modelling_Customer_churn.csv')
df.head()
```

# Standardizing data

```python
df_scaled = df.copy()
```

```python
columns = ['user_id','apm','income','height','reaction_time','age','kappa_ability','weight','amount','gregariousness']
```

```python
df1 = df_scaled[columns]
```

```python
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_ss= pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)
```

```python
df_ss.head()
```

```python
df_ss['user_id'] = df1['user_id']
```

```python
df_scaled.drop(columns=['apm','income','height','reaction_time','age','kappa_ability','weight','amount','gregariousness'], axis=1, inplace=True)
```

```python
df_scaled = df_scaled.merge(df_ss, how='inner', on='user_id')
```

```python
df_scaled.head()
```

```python
df_scaled.drop(columns=['user_id'], axis=1, inplace=True)
```

```python
X = df_scaled.drop(columns=['churn'], axis=1)
y = df['churn']
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
skfold = StratifiedKFold(n_splits=5)
skfold.get_n_splits(X, y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
scores =  cross_val_score(model, X, y, cv=skfold, scoring='recall')
```

```python
print(scores)
```

```python
print(np.mean(scores))
```

```python
from sklearn.model_selection import cross_val_predict
y_pred = cross_val_predict(model, X, y, cv=skfold)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, y_pred)
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y, y_pred))
```
